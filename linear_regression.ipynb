```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.uniform(-1, 1, (1000, 2))
x1 = data[:, 0]
y = data[:, 1]
x2 = x1**2

m = len(y)
epochs = 2
theta0 = 0.5
theta1 = 0.5
theta2 = 0.5
alpha = 0.7

for epoch in range(epochs):
  y_predict = theta0 + theta1*x1 + theta2*x2
  error = y_predict - y

  #Computing gradients
  dL_dtheta0 = (1/m) * np.sum(error)
  dL_dtheta1 = (1/m) * np.sum(error*x1)
  dL_dtheta2 = (1/m) * np.sum(error*x2)

  # Updating the parameters
  theta0 -= alpha * dL_dtheta0
  theta1 -= alpha * dL_dtheta1
  theta2 -= alpha * dL_dtheta2

  cost = (1/(2*m)) * np.sum(error**2)

  print(f"Epoch {epoch + 1}:")
  print(f"Cost: {cost:.4f}")
  print(f"theta0: {theta0:.4f}, theta1: {theta1:.4f}, theta2: {theta2:.4f}\n")

#Plotting
plt.figure(figsize=(8, 6))
plt.scatter(x1, y, label="Actual Data")
plt.plot(x1, y_predict, color="red", label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression demo")
plt.legend()
plt.show()
```
